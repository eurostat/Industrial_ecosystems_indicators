```python
# pandas
import pandas as pd 
pd.set_option('display.float_format', lambda x: '%.5f' % x)
weights = pd.read_excel("IES_Weights.xlsx")
weights
```

```python
import numpy as np
weights_table = pd.pivot_table(weights, values=['Weight'], index=['NACE2'],columns=["IES"], aggfunc=np.sum, fill_value=0)#reorde
weights_table 
```

```python
#addapting the weights to the NA classification

weights_table.rename(index={'C10': 'C10-C12', 'C13': 'C13-C15', 'D35': 'D', 'E37': 'E37-E39',
                           'C13': 'C13-C15', 'J59':'J59_J60','J62':'J62_J63',
                           'M74':'M74_M75','P85': 'P', 'Q87':'Q87_Q88', 'R90':'R90-R92'}, inplace=True)
```

```python
weights_table
```

```python
nacedrop=['C11','C12','C14','C15','E38','E39','J60','J63','M70','M75','Q88','R91','R92']
weights_table=weights_table.drop(nacedrop)
weights_table
```

```python
weights_table.to_excel('weights_table.xlsx')
```
