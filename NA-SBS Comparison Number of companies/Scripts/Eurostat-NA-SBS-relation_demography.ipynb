```python
import PySimpleGUI as sg
def selection(option):
    #sg.theme('DarkAmber')   # Add a touch of color
    #select the variable we want to calculate
    options = option

    # All the stuff inside your window.
    layout = [ 
            [sg.Text('Select one->'), sg.Listbox(options,select_mode=sg.LISTBOX_SELECT_MODE_SINGLE,size=(20,len(options)))],
            [sg.Button('Ok'), sg.Button('Cancel')]
        ]

    # Create the Window
    window = sg.Window('Make your choice', layout)

    # Event Loop to process "events" and get the "values" of the input
    while True:
        event, values = window.read()
        print( f"event={event}" )
        if event is None or event == 'Ok' or event == 'Cancel': # if user closes window or clicks cancel
            break
        
    # close  the window        
    window.close()

    if event == "Cancel":
        print( "You cancelled" )
    else:
        print('You entered ', values[0])
        sg.popup( f"You selected {values[0]}" )        
    return values
```

```python
# pandas
import pandas as pd 
# pip install eurostat  #herramienta para importar datos de eurostat
import eurostat
df_sbs = eurostat.get_data_df('sbs_sc_sca_r2', flags=False)
df_sbs
```

```python
#eliminate columns for years previous to 2011
years =[]
year_start=2005
year_study=2018
for y in range(year_start,year_study): 
    years.append(y)
years.append(2019)
years.append(2020)
df_sbs=df_sbs.drop(years, axis=1)#quitamos todas las columnas de añosque no nos interesan
#rename geo time column to be easier to handle because of the name
df_sbs.rename(columns={"geo\\time": "geo"}, inplace=True)
#We can get the unique values of the different variables
```

```python
nace_dif=set(df.nace_r2.unique())-set(df_sbs.nace_r2.unique())
```

```python
nace_sel=set(df.nace_r2.unique())-set(nace_dif)
nace_sel
```

```python
sizeclas=['TOTAL']
indic_sb=['V11910']
country=['EU27_2020']

#we select the rows that have the right unit, nace codes, items and countries
df_selection=df[df.nace_r2.isin(nace_sel)&df.sizeclas.isin(sizeclas)&df.indic_sb.isin(indic_sb)&df.geo.isin(country)]
df_selection.set_index('nace_r2', inplace=True)
df_selection
```

```python
df_selection=df_selection.drop(['indic_sb', 'sizeclas', 'geo'], axis=1)
```
