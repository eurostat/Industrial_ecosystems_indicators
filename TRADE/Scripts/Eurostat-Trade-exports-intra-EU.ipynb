```python
# we are going to sum the values for all the countries and reorganize so columns are years and rows are naces
#list of countries which are columns for the iteration
cols=countries_EU27

#we set the decimal to two numbers

pd.set_option('display.float_format', lambda x: '%.2f' % x)
import numpy as np

#we create an empty dataframe for the results of the aggregation of the countries
results_df = pd.DataFrame()

#loop to calculate values
year_i=2012
year_f=2021
#we transpose the values for the year selected
for i in range (year_i,year_f):
    df_target = pd.pivot_table(df_selection, values=i, index=['nace_r2'],columns=["geo"], aggfunc=np.sum, fill_value=0)#reorde
    df_target['sum'] = df_target[cols].sum(axis=1)#sums the columns of the countries to "sum" 
    #print(df_target)
    #save to file with the values for each year
    filename = "%s.xlsx" % str(variable_of_interest + '-year-'+ str(i))
    df_target.to_excel(filename)
    column_name=i
    results_df[column_name]=df_target['sum']#creates a new column with the name of the year and with the aggregated values for all the countries
    df_target=df_target.drop(cols, axis=1)#we eliminate the columns of the countries
results_df
```

```python
#create a dataframe for the results
results_compiled = pd.DataFrame()

import numpy as np
#loop to calculate values for years
for j in range (2012,2021):
    #for each year we create a base dataframe with the ecosystem weights
    weights_target = pd.read_excel("ecosystemweightsextended.xlsx")
    weights_target.set_index('NACE_R2', inplace=True)#we set nace as the index
    weights_target[variable_of_interest]=weights_target.index.map(results_df_naceweights[j])
    #we look in results_df_naceweights for the same nace as in weights_target and copy the
    # value of results_df_naceweights in a new column in the correspondig row (same nace)
    #so now we have the ecosystems columns and an extra column with the value of the variable of interest
    #the rows of the dataframe are the differnt naces in ecosystemweightsextended
   
    #create a results dataframe
    results=weights_target
    #we are going to iterate in the dataframe to multiply the value of the variable of interest by
    #the weight, so we are distributing the value among the ecosystems
    #once we finish we calculate the total as the sum of all the values aggregated in a ecosystem
    for i in range(0,14):
        results.iloc[:, i]=weights_target.iloc[:, i]*weights_target[variable_of_interest]#we multiply the total value added by the weight for each nace
    results.loc['total'] = results.select_dtypes(pd.np.number).sum() #we add the values of each colum in a row call total
    colname= str(variable_of_interest)+str(j)
    #we copy the results of the totals of the corresponding year to a new database that compiles all the years
    results_compiled[colname]=results.loc['total']
results_compiled = results_compiled.drop(variable_of_interest)
results_compiled  
```
