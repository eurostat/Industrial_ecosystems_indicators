```python
# pandas
import pandas as pd 
```

```python
value_added = pd.read_excel("Value added, gross.xlsx", index_col=0)
#set index in column 0
value_added 
```

```python
size_ratios_value_added = pd.read_excel("ratioSMEs- Value added at factor cost - million euro.xlsx", index_col=0)
size_ratios_value_added 
```

```python
#we create an empty dataframe to filter the ratios only for big companies
ratio_big = pd.DataFrame()
#set the first column to copy on the new dataframe
col=4
for i in range(0,8):
    add_column = size_ratios_value_added .iloc[:, col] #the column to copy
    ratio_big = pd.concat([ratio_big,add_column], axis = 1) # we concatenate the column to the new dataframe
    col=col+6 #we look for the next column to copy
ratio_big
```

```python
#create a dataframe for the results based on ratio dataframe
ratio_SME = ratio_big.copy()
for i in range(0,8):
    ratio_SME.iloc[:, i]=(1-ratio_big.iloc[:, i])
ratio_SME 
```

```python
#create a dataframe for the results based on ratio dataframe
results_BIG = value_added.copy()
for i in range(0,8):
    results_BIG.iloc[:, i]=ratio_big.iloc[:, i]*value_added .iloc[:, i]#we multiply the  total value added by ratio for big companies
results_BIG=results_BIG.add_suffix('_BIG')
results_BIG
```

```python
#save to file
fileratio='Value_added-BIG companies'
filename = "%s.xlsx" % fileratio
results_BIG.to_excel(filename)
```

```python
#create a dataframe for the results based on ratio dataframe
results_SME= value_added.copy()
#iterate to calculate the ratio times the value_added
for i in range(0,8):
    results_SME.iloc[:, i]=ratio_SME.iloc[:, i]*value_added.iloc[:, i]#we multiply the  total value added by ratio for SME companies
results_SME=results_SME.add_suffix('_SME')
results_SME
```

```python
#save to file
fileratio='Value_added-SME companies'
filename = "%s.xlsx" % fileratio
results_SME.to_excel(filename)
```

```python
#check that the results for big+SME= total results from value_added file
results= value_added.copy()
results = results_SME
for i in range(0,8):
    results.iloc[:, i]=results_SME.iloc[:, i]+results_BIG.iloc[:, i]
results
```
